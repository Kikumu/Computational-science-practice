```python
import numpy as np
import matplotlib.pyplot as plt
import math
#"step" function by robot

def func(x,U):
    #-2x+U
    val =(-2*x)+U
    return val
```

```python
a = 0
b = 0

t = a
x = b
h = 0.512 #step size
U_array = []
t_array = []
x_array = []
x_exact_array = []
while(t <= 15):
    if(t < 0 or t <= 5):
      U = 2
    elif(t > 5 or t <= 10):
      U = 1
    elif(t > 10 or t <= 15):
      U = 3
    value = func(U,x)
    U_array.append(U)
    k = h * value
    x = x + k
    t = t + h
    #value under graph
    x_exact = math.exp(-2*t)
    t_array.append(t)
    x_array.append(x)
    x_exact_array.append(x_exact)
    print(x,x_exact)
```

```python
y_axis = x_exact_array
x_axis = t_array
plt.plot(x_axis,y_axis)
plt.show()
```

```python
y_axis = U_array
x_axis = t_array
plt.plot(x_axis,y_axis)
plt.show()
```
