```python
import numpy as np
import matplotlib.pyplot as plt
#"step" function by robot

def func(x,U):
    #-2x+U
    val =(-2*x)+U
    return val
```

```python
a = 5
b = 7

t = a
x = b
h = 0.009 #step size
U_array = []
t_array = []
while(t <= 15):
    if(t < 0 or t <= 5):
      U = 2
    elif(t > 5 or t <= 10):
      U = 1
    elif(t > 10 or t <= 15):
      U = 3
    value = func(U,x)
    U_array.append(U)
    k = h * value
    x = x + k
    t = t + h
    t_array.append(t)
    print(x,t)
```

```python
y_axis = U_array
x_axis = t_array
plt.plot(x_axis,y_axis)
plt.show()
```
